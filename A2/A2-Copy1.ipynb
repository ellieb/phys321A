```python
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.ticker as tck
import sympy as sp
import math as ma

def plot_pendulum(th0):
    om_0 = 4*ma.pi
    tmax = 5
    t0 = 0
    dt = 0.001
    
    def step(thi, thj, t, dt): #i = i, j = i-1
        th = -(dt**2)*(om_0**2)*(ma.sin(thi)) + 2*thi - thj
        return th
    
    def sm_ang(t):
        return th0*ma.cos(om_0*t)
    
    sm_ang_vec = np.vectorize(sm_ang)
    
    th1 = th0 - 2*(om_0**2)*ma.sin(th0)*dt**2
    t = [t0, t0+dt]
    th = [th0, th1]

    while t[-1] < tmax:
        th.append(step(th[-1], th[-2], t[-1], dt))
        t.append(t[-1] + dt)

    fig, ax = plt.subplots(figsize=(12,8))
    ax.yaxis.set_major_formatter(tck.FormatStrFormatter('%g $\pi$'))
    ax.yaxis.set_major_locator(tck.MultipleLocator(base=1.0))
    ax.set_xlabel("Time (s)")
    ax.set_ylabel("Angle")
    fig.suptitle("Motion of a Simple Pendulum", fontsize=16)
    ax.plot(t, th, label="$\\theta (t)$")
    ax.plot(t, sm_ang_vec(t), label="$\\theta (t)$ with small angle approx.")
    ax.legend()
    
#1a
plot_pendulum(ma.pi/4)
```

```python
#1b
plot_pendulum(ma.pi/3)
```

```python
#1c
plot_pendulum(ma.pi/2)
```
